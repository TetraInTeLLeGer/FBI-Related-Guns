## Exploratory Data Analysis 1

```python
state = Gun_Data['state'].unique()
```

```python
Gun_Data['month'] = pd.to_datetime(Gun_Data['month'])
Gun_Data['month']
```

```python
for x in state: #The maximum of the long guns and plotting each country
    plt.title(x)
    ifstate = Gun_Data[Gun_Data['state'] == x]
    plt.scatter(ifstate['month'], ifstate['long_gun'], cmap = 'viridis')
    plt.xlabel('Years')
    plt.ylabel('Amount of Guns')
    maximum = ifstate['long_gun'].max()
    plt.grid()
    plt.show()
    print('The maximum of long guns in ', x ,'is',maximum)
```

```python
for x in state: #The maximum of the handguns and plotting each country
    plt.title(x)
    ifstate = Gun_Data[Gun_Data['state'] == x]
    plt.scatter(ifstate['month'], ifstate['handgun'], cmap = 'viridis')
    plt.xlabel('Years')
    plt.ylabel('Amount of Guns')
    maximum = ifstate['handgun'].max()
    plt.grid()
    plt.show()
    print('The maximum of handguns in ', x ,'is',maximum)
```

```python
themax = Gun_Data['long_gun'].idxmax() #Finding which city is higher in terms of numbers
themax2 = Gun_Data.iloc[themax]
print(themax2)
#Texas has the most guns in all of the state.
```

## Data Cleaning 2

```python
mask = (Gun_Data['month'] == '2016-07-01')
Gun_Data_16 = Gun_Data.loc[mask].set_index('state').reset_index()
US = US_poplution[US_poplution['Fact'].str.contains('2016').fillna(False)] #only for 2016 dataset
US = US.drop(['Fact Note'], axis = 1)
US1 = US.drop(['Fact'], axis = 1)
Gun_Data_16
```

```python
US = US.set_index('Fact')
```

```python
US = US.T
US
```

```python
US = US.reset_index().rename(columns = {'index': 'state'})
```

```python
US.iloc[:,1:] = US.iloc[:,1:].apply(lambda x: x.str.replace(',','').str.strip(' ').str.strip('$').str.replace('D','0').str.replace('FN','0').str.replace('Z','0').str.strip('-'))
```

```python
def per_remove(x):
    x[x.str.contains('%')]=x[x.str.contains('%')].str.strip('%').astype('float')/100
    return x
US = US.apply(per_remove)
US.info()
```

```python
US[US.columns[1:]] = US[US.columns[1:]].astype(float) #changing from string to float
US.info()
```

```python
US_Gun = US.merge(Gun_Data_16, how = 'left', on = 'state') #merging the 2 datasets
US_Gun
```

## Exploratory Data Analysis 2 

```python
empty = [] #finding the highest correlation between the hand gun and any type of capita
for x in US_Gun.columns[1:17]:
    example1 = np.corrcoef(US_Gun[x], US_Gun['handgun'])[0,1]
    empty.append(example1)
    fig, ax = plt.subplots(figsize = (9, 6))
    ax.scatter(US_Gun[x], US_Gun['handgun'], s=60, alpha=0.7, edgecolors="k")
    ax.set_xscale("log")
    ax.set_yscale('log')
    plt.show()
    print(x)
    print(example1)
print('The maximum correlation is Housing unit', max(empty))
#looks like the housing unit is highest correlation
```
